``Networkx`` is a python library designed to provide support to analysis of complex networks.

NetworkX is a Python package for the creation, manipulation, and study of the structure, dynamics, and functions of complex networks. It provides:

- tools for the study of the structure and dynamics of social, biological, and infrastructure networks;

- a standard programming interface and graph implementation that is suitable for many applications;

- a rapid development environment for collaborative, multidisciplinary projects;

- an interface to existing numerical algorithms and code written in C, C++, and FORTRAN; and

- the ability to painlessly work with large nonstandard data sets.

With NetworkX you can load and store networks in standard and nonstandard data formats, generate many types of random and classic networks, analyze network structure, build network models, design new network algorithms, draw networks, and much more.

### 2. Design our first graph

``networkx`` provides support for several graph models.

Among them:

- undirected graphs, available through the Graph class
- directed graphs, available through the DiGraph class

In this tutorial we will focus only on undirected graphs.

#### 2.1.1: We can build a graph adding nodes as follows:

```python
g = nx.Graph()
g.add_node('a')
g.add_node('b')
g.add_node('c')

fig = plt.figure(figsize=(2,2))
nx.draw(g ,with_labels=True)
```

#### 2.1.2: We can build a graph adding edges as follows:

```python
g1 = nx.Graph()
g1.add_edge('a', 'b')
nx.draw(g1 ,with_labels=True)
```

#### 2.1.3:  We can also add edges directly without adding the nodes

```python
g1 = nx.Graph()
g1.add_edge('a', 'b')
g1.add_edge('a', 'c')
g1.add_edge('a', 'd')

fig = plt.figure(figsize=(2,2))
nx.draw(g1 ,with_labels=True)
```

#### 2.1.4: Nodes and edges can be removed

```python
g1.remove_node("c")
g1.remove_edge("a", "d")

#re-draw the graph
fig = plt.figure(figsize=(2,2))
nx.draw(g1, with_labels=True)
```

#### Another way to add edges: add_edges_from()

```python
g2 = nx.Graph()
g2.add_edges_from(g1.edges)
#Note that: node d in g1 is ignored due to no edges
g2.add_edge('b', 'NN')
fig = plt.figure(figsize=(2,2))
nx.draw(g2 ,with_labels=True)
```

```python
g3 = nx.Graph()
# Note that: nodes will be added only, edges are not added
g3.add_nodes_from(g2)
fig = plt.figure(figsize=(2,2))
nx.draw(g3 ,with_labels=True)
```

#### 3.5: define a function draw_network(G, dict_centrality) to draw a network and the node size is based on the following specified centralities:

    nx.degree_centrality()
    nx.betweenness_centrality()
    nx.eigenvector_centrality()
    nx.harmonic_centrality()

```python
def draw_network(G, dict_centrality, pos):
    cc = dict_centrality
    if (max(cc.values()) < 1):
        adapted_cc_values = [ v **2 * 10000 for v in cc.values()]
    else:
        ## for harmonic centrality
        adapted_cc_values = [ v * 100 for v in cc.values()]
    nx.draw(G, nodelist=cc.keys(), node_size=adapted_cc_values, pos=pos, with_labels=True, font_color="white")
```
